## Run ``cpac``

First, we'll look at the usage guidelines:

````bash
cpac --help
````

We want to use Singularity for our container platform, so we'll use the ``--platform`` flag to set ``--platform singularity``.

We don't need FreeSurfer for our pipeline configuration, so we can use the smaller ``-lite`` variant of C-PAC. We'll use v1.8.4 with ``--tag release-v1.8.4-lite``.

Now we can look at the usage info for running with that platform and tag:

````bash
cpac \
  --platform singularity \
  --tag release-v1.8.4-lite \
  run --help
````

* There are three required positional arguments: `bids_dir`, `output_dir` and level of analysis. In this example, we'll use
    * a BIDS directory from S3 for the `bids_dir`,
    * a subdirectory of the current working directory for the `output_dir` (`./cpac_runs`), and
    * ``participant`` as the level of analysis.
* We want to use the ``rbc-options`` preconfiguration, so we'll set that with the ``--preconfig`` flag.
* We just want to run a single subject for this example, so we'll specify one with the ``--participant_label`` flag.
* We'll use the ``--n_cpus`` and ``--mem_gb`` flags to specify resource allocation

````bash
mkdir cpac_runs
cpac \
  --platform singularity \
  --tag release-v1.8.4-lite \
  run \
  s3://fcp-indi/data/Projects/ADHD200/RawDataBIDS/NYU cpac_runs participant \
  --preconfig rbc-options \
  --participant_label 0010042 \
  --n_cpus 4 \
  --mem_gb 20
````

* ``cpac`` found our ``C-PAC`` Singularity image and launched a container with the appropriate directory bindings, matching paths inside the container to those on my local system, and set the container user to the user who launched the container, i.e., me. Inside the container, it started a ``C-PAC`` run.
* ``C-PAC`` loaded the ``rbc-options`` preconfig which cascaded from the ``fx-options``, ``fmirprep-options`` and default preconfigured pipelines.
* ``C-PAC`` logged its configuration info and inside-the-image run command.
